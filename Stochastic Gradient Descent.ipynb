```python
import pandas as pd
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error
```

```python
from sklearn.model_selection import train_test_split
```

```python
X1= np.random.randint(1,10,100)
X2= np.random.randint(1,10,100)
```

```python
Y= 4*X1 + 10*X2
Y = np.vstack(Y)
```

```python
X1 = np.vstack(X1)
X2 = np.vstack(X2)
X = np.hstack((X1,X2))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=1)
```

```python
class linearRegression:
    def __init__(self, epochs=10000, l_Rate=0.0001, random_state=1):
        self.epochs = epochs
        self.learningRate = l_Rate
        self.randomState = random_state
    def compError(self, cf, X, Y):
        totalError = 0
        for i in range(0, len(X)):
            yhat = cf[0]
            for j in range(0,len(X[0])):
                yhat += cf[j + 1] * X[i][j]
            totalError += (Y[i] - yhat) ** 2
        return totalError/float(len(X))
    def yhat_predict(self, x_feat, coefficients):
        yhat = coefficients[0]
        for i in range(len(x_feat)):
            yhat += coefficients[i + 1] * x_feat[i]
        return yhat
    def grad_descent(self, X, Y, cf, l_rate):
        for i in range(len(X)):
            yhat = self.yhat_predict(X[i], cf)
            error = yhat - Y[i]
            cf[0] = cf[0] - l_rate * error
            for j in range(len(X[i])): 
                cf[j + 1] = cf[j + 1] - l_rate * error * X[i][j]
        return cf
    def fit(self, X, Y):
        iterations = self.epochs
        l_rate = self.learningRate
        cf = np.random.uniform(0,1,len(X[0])+1)
        for epoch in range(iterations):
            cf = self.grad_descent(X, Y, cf, l_rate)
            error = self.compError(cf, X, Y)
            print("Epoch: ", epoch, "           MSE: ", error)
        self.cf = cf
    def predict(self, X):
        Y = []
        coefficients = self.cf
        for i in range(len(X)):
            yhat = coefficients[0]
            for j in range(len(X[i])):
                yhat += coefficients[j + 1] * X[i][j]
            Y.append(yhat)
        Y = np.array(Y)
        return Y
```

```python
linearReg = linearRegression(epochs=10000,random_state=100)
```

```python
linearReg.fit(X_train,y_train)
```

```python
y_pred = linearReg.predict(X_test)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred) 
```

For pre-processing, we used tf-idf vectorizer.
It transforms a count matrix to a normalized tf or tf-idf representation.

Tf means term-frequency while tf-idf means term-frequency times inverse document-frequency. This is a common term weighting scheme in information retrieval, that has also found good use in document classification.

The goal of using tf-idf instead of the raw frequencies of occurrence of a token in a given document is to scale down the impact of tokens that occur very frequently in a given corpus and that are hence empirically less informative than features that occur in a small fraction of the training corpus.
